```python
import pandas as pd

# Import PSSA data 

pssa_2019 = pd.read_excel(r"data/raw data/2019 pssa school level data.xlsx")
```

```python
# View head of data 

pssa_2019.head()
```

```python
import pandas as pd


# Filter for English and Math, and "All Students"
pssa_2019_filter = pssa_2019[
    (pssa_2019["Subject"].isin(["English Language Arts", "Math"])) &
    (pssa_2019["Group"] == "All Students") &
    (pssa_2019["Grade"] == "Total")
]

# Group by district and calculate mean percent proficient
result_2019 = (
    pssa_2019_filter
    .groupby("District Name", as_index=False)["Percent Proficient"]
    .mean()
    .rename(columns={"Percent Proficient": "mean_percent_proficient_2019"})
)

result_2019.head()
```

```python
import pandas as pd

# Filter for English and Math, and "All Students"
pssa_2019_filter = pssa_2019[
    (pssa_2019["Subject"].isin(["English Language Arts", "Math"])) &
    (pssa_2019["Group"] == "All Students") &
    (pssa_2019["Grade"] == "Total")
]

# Group by district and subject, then calculate mean percent proficient
result_2019 = (
    pssa_2019_filter
    .groupby(["District Name", "Subject"], as_index=False)["Percent Proficient"]
    .mean()
    .rename(columns={"Percent Proficient": "mean_percent_proficient"})
)

# Pivot the data to create separate columns for English and Math
result_2019_pivot = result_2019.pivot(
    index="District Name",
    columns="Subject",
    values="mean_percent_proficient"
).reset_index()

# Rename columns for clarity
result_2019_pivot = result_2019_pivot.rename(
    columns={
        "English Language Arts": "mean_proficient_english_2019",
        "Math": "mean_proficient_math_2019"
    }
)

result_2019_pivot.head()
```
