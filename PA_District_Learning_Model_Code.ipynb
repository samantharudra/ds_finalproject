# PA District Learning Models Data Code 

### Import Packages

```python
import pandas as pd
import numpy as np
```

### Import CSV

```python
pa_district_data = pd.read_csv("Pennsylvania_Districts_LearningModelData_Final.csv")
pa_district_data.head()
```

### Clean Data

```python
PA_district_learning_models_clean = (
    pa_district_data[['DistrictName', 'DistrictType', 'LearningModel']]
    .fillna ({'LearningModel': 'None'})
) #select columns and replace missing values in the LearningModel column

PA_district_learning_models_clean = (
    PA_district_learning_models_clean
    .groupby(['DistrictName', 'DistrictType'], as_index=False)
    .agg(
        Virtual=('LearningModel', lambda x: (x == 'Virtual').sum()),
        Hybrid=('LearningModel', lambda x: (x == 'Hybrid').sum()),
        In_Person=('LearningModel', lambda x: (x == 'In-person').sum())
    )
)
PA_district_learning_models_clean.head()
PA_district_learning_models_clean.to_csv('PA_district_learning_models_clean.csv', index=False)
```
