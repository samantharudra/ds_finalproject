# Merging Data

```python
import pandas as pd
import numpy as np
```

### Import Data

```python
learning_models = pd.read_csv("data/cleaned data/PA_district_learning_models_clean.csv")
learning_models.rename(columns={'DistrictName': 'District Name'}, inplace=True)
learning_models["District Name"] = learning_models["District Name"].str.replace("SD", "School District")
learning_models["District Name"] = learning_models["District Name"].str.upper()
learning_models.head()
```

```python
test_scores = pd.read_excel("data/cleaned data/district_test_data.xlsx")
test_scores["District Name"] = test_scores["District Name"].str.replace("SD", "SCHOOL DISTRICT")
test_scores.head()
```

```python
district_data = pd.read_csv("district_data.csv")
district_data.rename(columns ={'name':"District Name"}, inplace = True)
district_data["District Name"] = district_data["District Name"].str.upper()
district_data.head()
```

### Merge Data

```python
merged_data = learning_models.merge(test_scores, on='District Name', how='outer') \
               .merge(district_data, on='District Name', how='outer')
merged_data.head()
```

```python
merged_data.to_csv('merged_data.csv', index=False)
```
